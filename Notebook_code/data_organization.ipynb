```python
#importing libraries
import pandas as pd
import os
import csv
import json
import numpy as np

```

```python
#importing first csv file
reviews_df = pd.read_csv("../Yelp_data/yelp_reviews.csv")
reviews_df.head()
```

```python
#importing 2nd csv file
inspections_df = pd.read_csv("../Yelp_data/restaurant_inspections.csv")
inspections_df.head()
```

```python
#changing name column to lowercase to match other df
reviews_df['name'] = reviews_df['name'].str.lower()
reviews_df.head()
```

```python
#duplicating df
reviews_cleaning_df = reviews_df
reviews_cleaning_df.head()
```

```python
#dropping phone column
reviews_cleaning_df = reviews_cleaning_df.drop(['phone'], axis = 1)
reviews_cleaning_df.head()
```

```python
#printing 2nd df
inspections_df.head()
```

```python
#dropping unneeded columns
inspections_cleaned_df = inspections_df.drop(['business_id', 'violation_id', 'violation_description', 'risk_category', 'phone'], axis = 1)
inspections_cleaned_df.head()
```

```python
#renaming business city to city to match Reviews df
inspections_cleaned_df.rename(columns = {'business_city':'city'}, inplace = True)
inspections_cleaned_df.head()
```

```python
#renaming business_state to state to match other df
inspections_cleaned_df.rename(columns = {'business_state':'state'}, inplace = True)
inspections_cleaned_df.head()
```

```python
#dropping null values from Inspection df
inspections_cleaned_df = inspections_cleaned_df.dropna(how='any')
inspections_cleaned_df.head()
```

```python
#checking number of values
inspections_cleaned_df.info()
```

```python
#checking individual unique values
inspections_cleaned_df.value_counts()
```

```python
#double checking values from inspection type column
inspections_cleaned_df['inspection_type'].value_counts()
```

```python
#sorting inspection date to ensure we have newer values up top
inspections_cleaned_df.sort_values(by=['inspection_date'], inplace=True, ascending=True)
inspections_cleaned_df.head()
```

```python
#keeping first values of the name column, dropping any duplicates
inspections_cleaned_df.drop_duplicates(subset=['name'], keep='first', inplace=True, ignore_index=False)
inspections_cleaned_df.head()
```

```python
#checking values
inspections_cleaned_df.info()
```

```python
#checking if other value is removed from type
inspections_cleaned_df['inspection_type'].value_counts()
```

```python
#making new clean df and resetting index
ins_clean_df = inspections_cleaned_df[['name', 'address', 'city', 'state', 'zip', 'inspection_score']]
ins_clean_df = ins_clean_df.reset_index(drop=True)
ins_clean_df['zip'] = ins_clean_df['zip'].astype(str)
ins_clean_df.head()
```

```python
ins_clean_df.info()
```

```python
#working on reviews df
reviews_cleaning_df.head()
```

```python
#reorder df
reviews_cleaning_df = reviews_cleaning_df[['name', 'address', 'city', 'state', 'zip_code', 'rating', 'reviews']]
reviews_cleaning_df.head()
```

```python
reviews_cleaning_df['state'].value_counts()
```

```python
reviews_cleaning_df['state'] = reviews_cleaning_df['state'].replace({'X': 'CA'})
reviews_cleaning_df['state'].value_counts()
```

```python
reviews_cleaning_df = reviews_cleaning_df.drop(1)
reviews_cleaning_df.head()
```

```python
reviews_cleaning_df = reviews_cleaning_df.reset_index(drop=True)
reviews_cleaning_df.head()
```

```python
reviews_cleaning_df['zip_code'].value_counts()
```

```python
reviews_cleaning_df = reviews_cleaning_df.dropna(how='any')
```

```python
reviews_cleaning_df['zip_code'] = reviews_cleaning_df['zip_code'].astype(int)
reviews_cleaning_df.head()
```

```python
reviews_cleaning_df['zip_code'] = reviews_cleaning_df['zip_code'].astype(str)
reviews_cleaning_df.head()
```

```python
reviews_cleaning_df.info()
```
